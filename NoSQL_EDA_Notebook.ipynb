```python
# EDA - review of uk_food - establishments
# import packages
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# mongo client instance
mongo = MongoClient(port=27017)
# assign db and collections to var
uk_food = mongo['uk_food']
establishments = uk_food['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene':20}
# Use count_documents to display the number of documents in the result
print(f"Number of documents in result: ", establishments.count_documents(query))
print(f"-"*45)
# Display the first document in the results using pprint
pprint(establishments.find_one())
```

```python
# Convert the result to a Pandas DataFrame
Hyg_20 = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
print(f"Number of rows in df: ", len(Hyg_20))
# Display the first 10 rows of the DataFrame
Hyg_20.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': 4}}
# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"London has {count} establishments with a RV greater than/equal to 4")
print(f"-"*45)
# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
lon_rv4 = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
print(f"Number of rows in df: ", len(lon_rv4))
# Display the first 10 rows of the DataFrame
lon_rv4.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# coords for Penang Flavours
print(establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.latitude', 'geocode.longitude'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000
# concat query
query = {'RatingValue': 5, 
         'geocode.latitude': {'$lte': (latitude + degree_search), '$gte': (latitude - degree_search)}, 
         'geocode.longitude': {'$lte': (longitude + degree_search), '$gte': (longitude - degree_search)} 
         }
sort =  sort = [('scores.Hygiene', 1)] # asc for now
fields = {'BusinessName':1, 'BusinessType':1, 'RatingValue':1, 'scores.Hygiene':1, 'geocode.latitude':1, 'geocode.longitude':1}
#limit = 10
# Print the results
results = list(establishments.find(query, fields).sort(sort))
pprint(results)
```

```python
# Convert result to Pandas DataFrame
pf_t5_lowhyg = pd.DataFrame(results)
pf_t5_lowhyg.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match = {'$match': {'scores.Hygiene': 0}}
# 2. Groups the matches by Local Authority
group = {'$group': {'_id': '$LocalAuthorityName', 'count': { '$sum': 1 }}}
# 3. Sorts the matches from highest to lowest
sort = {'$sort': {'count': -1 }}

pipeline = [match, group, sort]
results = list(establishments.aggregate(pipeline))
# Print the number of documents in the result
print(f"# of docs in result: ", len(results))
print(f"-"*45)
# Print the first 10 results
pprint(results[:10])
```

```python
# Convert the result to a Pandas DataFrame
pipe_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f"Number of rows in df: ", len(pipe_df))
# Display the first 10 rows of the DataFrame
pipe_df.head(10)
```
